# Drift corrections

Although no drift was observed in the coulometry counts throughout the analysis period as a whole, there are a number of DIC cells that seem to have linearly drifting values through the analysis day. We can apply a drift correction and compare those values to the uncorrected values. If significantly different, the drift corrected values will be used. 

For Laurel, drift possibly occurred on cell numbers 6 and 10. 

For Lucy, drift likely occurred on cell numbers 4 and 8, and possibly alsoon 10, 13, and 15. 

A drift corrected value will only be determined for given cells. 

The impact of the drift correction on the sample values is assessed in `SD035_DICTA_samples_QC.ipynb`.

```python
piccolo_laurel.df.loc[piccolo_laurel.df["Cast"].isin(casts_for_ian)].head(3)
piccolo_lucy.df.loc[piccolo_lucy.df["Cast"].isin(casts_for_ian)].head(3)
```

```python
fig = plt.figure(figsize = (5, 7))
ax0 = fig.add_subplot(111)

for i, c in zip(casts_for_ian, ["b", "orange", "r", "g", "y", "grey"]): 
    cond = piccolo_laurel.df["Cast"] == i
    ax0.scatter(piccolo_laurel.df["dic_sample_umol_per_kg"].loc[cond], piccolo_laurel.df["Depth (m)"].loc[cond], label = "Cast " + str(i), c = c)
    cond = piccolo_lucy.df["Cast"] == i
    ax0.scatter(piccolo_lucy.df["dic_sample_umol_per_kg"].loc[cond], piccolo_lucy.df["Depth (m)"].loc[cond], label = "Cast " + str(i), c = c)

ax0.set_ylim(600,0)

ax0.legend()

fig.tight_layout();
```

```python
df_dic_for_ian = pd.concat([piccolo_laurel.df.loc[piccolo_laurel.df["Cast"].isin(casts_for_ian)], piccolo_lucy.df.loc[piccolo_lucy.df["Cast"].isin(casts_for_ian)]], ignore_index = True)
df_dic_for_ian.loc[~df_dic_for_ian["Station"].isin([8888, 2222, 3333, 9999])].reset_index(drop = True, inplace = True)


fdir = "../ED_PICCOLO_output_data/ED_PICCOLO_vindta_processed/prelim_data_for_Ian_and_Ruth/"
fname = "SD035_dic_prelim_forIanandRuth_" + datetime.date.today().strftime(format = "%Y%m%d") + ".csv"
# df_dic_for_ian.to_csv(fdir + fname, index = False)
```

```python
fig = plt.figure(figsize = (15, 10))
ax0 = fig.add_subplot(231)
ax1 = fig.add_subplot(232)
ax2 = fig.add_subplot(233)
ax3 = fig.add_subplot(234)
ax4 = fig.add_subplot(235)
ax5 = fig.add_subplot(236)

for ax, cast in zip([ax0, ax1, ax2, ax3, ax4, ax5], casts_for_ian): 

    cond = df_dic_for_ian["Cast"] == cast

    ax.scatter(df_dic_for_ian["dic_sample_umol_per_kg"].loc[cond], df_dic_for_ian["Depth (m)"].loc[cond], c = "grey", edgecolor = "k", s = 100)

    ax.text(0.03, 0.05, "Cast " + str(cast), transform = ax.transAxes, fontsize = 14, weight = "bold")
    ax.set_ylim(600, 0)
    ax.set_xlim(2000, 2300)
    ax.set_ylabel("Depth [m]", fontsize = 14)
    ax.set_xlabel("DIC [$\mu$mol kg$^{-1}$]", fontsize = 14)
    ax.tick_params(labelsize = 14, pad = 10)
    ax.grid()
    

fig.tight_layout();

fdir = "../ED_PICCOLO_output_data/ED_PICCOLO_vindta_processed/prelim_data_for_Ian_and_Ruth/"
fname = "SD035_dic_prelim_forIanandRuth_" + datetime.date.today().strftime(format = "%Y%m%d") + ".png";
# fig.savefig(fdir + fname, bbox_inches = "tight");
```
